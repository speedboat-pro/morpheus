```r
trips_df <- SparkR::sql("SELECT * FROM samples.nyctaxi.trips")
```

```r
windowSpec <- windowPartitionBy(column("pickup_zip"))
windowSpec <- orderBy(windowSpec, column("fare_amount"))

# Apply row_number window function
df_with_row_num <- withColumn(trips_df, "row_number", over(row_number(), windowSpec))

# Example with descending order
windowSpec_desc <- windowPartitionBy(column("pickup_zip"))
windowSpec_desc <- orderBy(windowSpec_desc, desc(column("fare_amount")))

df_with_row_num_desc <- withColumn(trips_df, "row_number", over(row_number(), windowSpec_desc))
```
