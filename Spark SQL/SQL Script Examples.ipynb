## Stored Procedures

```sql
CREATE OR REPLACE PROCEDURE add(x INT, y INT, OUT sum INT, INOUT total INT)
    LANGUAGE SQL
    SQL SECURITY INVOKER
    COMMENT 'Add two numbers'
    AS BEGIN
        SET sum = x + y;
        SET total = total + sum;
    END;
```

```sql
DECLARE sum INT;
DECLARE total INT DEFAULT 0;
CALL add(1, 2, sum, total);
SELECT sum, total;
```

```sql
CALL add(3, 4, sum, total);
```

```sql
SELECT sum, total;
```

```sql
-- The last executed query is the result set of a procedure
CREATE PROCEDURE greeting_sample(IN mode STRING COMMENT 'informal or formal')
    LANGUAGE SQL
    SQL SECURITY INVOKER
    AS BEGIN
        SELECT 'Hello!';
        CASE mode WHEN 'informal' THEN SELECT 'Hi!';
                  WHEN 'formal' THEN SELECT 'Pleased to meet you.';
        END CASE;
    END;
```

```sql
CALL greeting_sample('informal');
```

```sql
CALL greeting_sample('formal');
```

```sql
CALL greeting_sample('casual');
```
