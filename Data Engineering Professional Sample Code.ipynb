````sql
-- DBU consumption by product this month
SELECT
  billing_origin_product,
  usage_date,
  sum(usage_quantity) as usage_quantity
FROM system.billing.usage
WHERE
  month(usage_date) = month(NOW())
  AND year(usage_date) = year(NOW())
GROUP BY billing_origin_product, usage_date

-- Attribute costs to compute resource owners
SELECT
  u.record_id,
  c.cluster_id,
  max_by(c.owned_by, c.change_time) owned_by,
  any_value(u.usage_quantity) usage_quantity
FROM system.billing.usage u
JOIN system.compute.clusters c
WHERE u.usage_metadata.cluster_id = c.cluster_id
GROUP BY 1, 2
````

````sql
-- Which warehouses are actively running?
SELECT
  we.warehouse_id,
  we.event_time,
  TIMESTAMPDIFF(MINUTE, we.event_time, NOW()) / 60.0 AS running_hours,
  we.cluster_count
FROM system.compute.warehouse_events we
WHERE we.event_type = 'RUNNING'
AND NOT EXISTS (
  SELECT 1
  FROM system.compute.warehouse_events we2
  WHERE we2.warehouse_id = we.warehouse_id
  AND we2.event_time > we.event_time
)

-- Warehouses with most uptime in a month
SELECT
  warehouse_id,
  SUM(TIMESTAMPDIFF(MINUTE, start_time, end_time)) / 60.0 AS uptime_hours
FROM (
  SELECT
    starting.warehouse_id,
    starting.event_time AS start_time,
    MIN(stopping.event_time) AS end_time
  FROM system.compute.warehouse_events AS starting
  JOIN system.compute.warehouse_events AS stopping
  WHERE stopping.event_type = 'STOPPED'
  AND starting.event_type = 'STARTING'
  GROUP BY starting.warehouse_id, starting.event_time
)
GROUP BY warehouse_id
ORDER BY uptime_hours DESC
````

````sql
WITH last_status_per_update AS (
  SELECT
    origin.pipeline_id,
    origin.pipeline_name,
    origin.update_id AS pipeline_update_id,
    FROM_JSON(details, 'struct<...>').update_progress.state AS last_update_state,
    timestamp,
    ROW_NUMBER() OVER (PARTITION BY origin.update_id ORDER BY timestamp DESC) AS rn
  FROM event_log_raw
  WHERE event_type = 'update_progress'
  QUALIFY rn = 1
)
SELECT
  pipeline_id,
  pipeline_name,
  pipeline_update_id,
  start_time,
  end_time,
  duration_seconds,
  pipeline_update_status
FROM last_status_per_update
ORDER BY start_time DESC;
````

````sql
CREATE OR REPLACE TEMP VIEW event_log_raw
AS SELECT * FROM delta.`<event-log-path>`;

-- Get latest update ID
CREATE OR REPLACE TEMP VIEW latest_update AS
SELECT origin.update_id AS id
FROM event_log_raw
WHERE event_type = 'create_update'
ORDER BY timestamp DESC
LIMIT 1;
````

````sql
-- View events on a materialized view
SELECT timestamp, message, details
FROM event_log(table(my_mv))
WHERE level in ('INFO', 'WARN', 'ERROR')
ORDER BY timestamp;

-- Query lineage information
SELECT
  details:flow_definition.output_dataset as output_dataset,
  details:flow_definition.input_datasets as input_dataset
FROM event_log(''), latest_update
WHERE event_type = 'flow_definition' 
  AND origin.update_id = latest_update.id;

-- Query data quality expectations
WITH expectations_parsed AS (
  SELECT explode(
    from_json(
      details:flow_progress.data_quality.expectations,
      "array<struct<name:string,dataset:string,passed_records:int,failed_records:int>>"
    )
  ) row_expectations
  FROM event_log(table(my_st)), latest_update
  WHERE event_type = 'flow_progress'
    AND origin.update_id = latest_update.id
)
SELECT
  row_expectations.dataset as dataset,
  row_expectations.name as expectation,
  SUM(row_expectations.passed_records) as passing_records,
  SUM(row_expectations.failed_records) as failing_records
FROM expectations_parsed
GROUP BY row_expectations.dataset, row_expectations.name;
````

````sql
WITH last_status_per_update AS (
  SELECT
    origin.pipeline_id AS pipeline_id,
    origin.pipeline_name AS pipeline_name,
    origin.update_id AS pipeline_update_id,
    FROM_JSON(details, 'struct<update_progress:struct<state:string>>').update_progress.state AS last_update_state,
    timestamp,
    ROW_NUMBER() OVER (PARTITION BY origin.update_id ORDER BY timestamp DESC) AS rn
  FROM event_log_raw
  WHERE event_type = 'update_progress'
  QUALIFY rn = 1
),
update_durations AS (
  SELECT
    origin.pipeline_id AS pipeline_id,
    origin.pipeline_name AS pipeline_name,
    origin.update_id AS pipeline_update_id,
    MIN(CASE WHEN event_type = 'create_update' THEN timestamp END) AS start_time,
    COALESCE(
      MAX(CASE
        WHEN event_type = 'update_progress'
        AND FROM_JSON(details, 'struct<update_progress:struct<state:string>>').update_progress.state 
            IN ('COMPLETED', 'FAILED', 'CANCELED')
        THEN timestamp
      END),
      current_timestamp()
    ) AS end_time
  FROM event_log_raw
  WHERE event_type IN ('create_update', 'update_progress')
    AND origin.update_id IS NOT NULL
  GROUP BY pipeline_id, pipeline_name, pipeline_update_id
  HAVING start_time IS NOT NULL
)
SELECT
  s.pipeline_id,
  s.pipeline_name,
  s.pipeline_update_id,
  d.start_time,
  d.end_time,
  CASE
    WHEN d.start_time IS NOT NULL AND d.end_time IS NOT NULL THEN
      ROUND(TIMESTAMPDIFF(MILLISECOND, d.start_time, d.end_time) / 1000)
    ELSE NULL
  END AS duration_seconds,
  s.last_update_state AS pipeline_update_status
FROM last_status_per_update s
JOIN update_durations d
  ON s.pipeline_id = d.pipeline_id
  AND s.pipeline_update_id = d.pipeline_update_id
ORDER BY d.start_time DESC;
````

````sql
WITH flow_progress_raw AS (
  SELECT
    origin.pipeline_name AS pipeline_name,
    origin.pipeline_id AS pipeline_id,
    origin.flow_name AS table_name,
    origin.update_id AS update_id,
    timestamp,
    details:flow_progress.status AS status,
    TRY_CAST(details:flow_progress.metrics.num_output_rows AS BIGINT) AS num_output_rows,
    TRY_CAST(details:flow_progress.metrics.num_upserted_rows AS BIGINT) AS num_upserted_rows,
    TRY_CAST(details:flow_progress.metrics.num_deleted_rows AS BIGINT) AS num_deleted_rows,
    TRY_CAST(details:flow_progress.data_quality.dropped_records AS BIGINT) AS num_expectation_dropped_rows,
    FROM_JSON(
      details:flow_progress.data_quality.expectations,
      SCHEMA_OF_JSON("[{'name':'str', 'dataset':'str', 'passed_records':42, 'failed_records':42}]")
    ) AS expectations_array
  FROM event_log_raw
  WHERE event_type = 'flow_progress'
    AND origin.flow_name IS NOT NULL
    AND origin.flow_name != 'pipelines.flowTimeMetrics.missingFlowName'
),
aggregated_flows AS (
  SELECT
    pipeline_name,
    pipeline_id,
    update_id,
    table_name,
    MIN(CASE WHEN status IN ('STARTING', 'RUNNING', 'COMPLETED') THEN timestamp END) AS start_timestamp,
    MAX(CASE WHEN status IN ('STARTING', 'RUNNING', 'COMPLETED') THEN timestamp END) AS end_timestamp,
    MAX_BY(status, timestamp) FILTER (
      WHERE status IN ('COMPLETED', 'FAILED', 'CANCELLED', 'EXCLUDED', 'SKIPPED', 'STOPPED', 'IDLE')
    ) AS final_status,
    SUM(COALESCE(num_output_rows, 0)) AS total_output_records,
    SUM(COALESCE(num_upserted_rows, 0)) AS total_upserted_records,
    SUM(COALESCE(num_deleted_rows, 0)) AS total_deleted_records,
    MAX(COALESCE(num_expectation_dropped_rows, 0)) AS total_expectation_dropped_records,
    MAX(expectations_array) AS total_expectations
  FROM flow_progress_raw
  GROUP BY pipeline_name, pipeline_id, update_id, table_name
)
SELECT
  af.pipeline_name,
  af.pipeline_id,
  af.update_id,
  af.table_name,
  af.start_timestamp,
  af.end_timestamp,
  af.final_status,
  CASE
    WHEN af.start_timestamp IS NOT NULL AND af.end_timestamp IS NOT NULL THEN
      ROUND(TIMESTAMPDIFF(MILLISECOND, af.start_timestamp, af.end_timestamp) / 1000)
    ELSE NULL
  END AS duration_seconds,
  af.total_output_records,
  af.total_upserted_records,
  af.total_deleted_records,
  af.total_expectation_dropped_records,
  af.total_expectations
FROM aggregated_flows af
-- Optional: filter to latest update only
WHERE af.update_id = (
  SELECT update_id
  FROM aggregated_flows
  ORDER BY end_timestamp DESC
  LIMIT 1
)
ORDER BY af.end_timestamp DESC, af.pipeline_name, af.pipeline_id, af.update_id, af.table_name;
````

````sql
WITH rounded_data AS (
  SELECT
    DATE_TRUNC('HOUR', event_time) AS evaluated_at,
    CONCAT(catalog_name, '.', schema_name, '.', table_name) AS full_table_name,
    status,
    MAX(downstream_impact.num_queries_on_affected_tables) AS impacted_queries,
    MAX(freshness.commit_freshness.predicted_value) AS commit_expected,
    MAX(freshness.commit_freshness.last_value) AS commit_actual,
    MAX(completeness.daily_row_count.min_predicted_value) AS completeness_expected,
    MAX(completeness.daily_row_count.last_value) AS completeness_actual
  FROM system.data_quality_monitoring.table_results
  GROUP BY ALL
)
SELECT
  evaluated_at,
  full_table_name,
  status,
  commit_expected,
  commit_actual,
  completeness_expected,
  completeness_actual,
  impacted_queries
FROM rounded_data
WHERE
  evaluated_at >= current_timestamp() - INTERVAL 6 HOURS
  AND impacted_queries > :min_tables_affected
  AND status = 'Unhealthy';
````

````sql
-- Create mapping table
DROP TABLE IF EXISTS valid_users;
CREATE TABLE valid_users(username string);
INSERT INTO valid_users VALUES
  ('fred@databricks.com'),
  ('barney@databricks.com');

-- Create filter function
CREATE FUNCTION row_filter()
RETURN EXISTS(
  SELECT 1 FROM valid_users v
  WHERE v.username = CURRENT_USER()
);

-- Create table with row filter
CREATE TABLE data_table (x INT, y INT, z INT)
WITH ROW FILTER row_filter ON ();

INSERT INTO data_table VALUES (1, 2, 3), (4, 5, 6), (7, 8, 9);

-- Query returns data only if user is in valid_users table
SELECT * FROM data_table;
````

### **What are Dynamic Views?**

Dynamic views allow you to **define logic over one or more tables using SQL**. They are read-only and useful for:
* **Applying transformation logic** - Joins, CASE statements
* **Sharing filtered data** - Using Delta Sharing (where ABAC is not supported)
* **Fine-grained access control** - Security at column or row level
* **Data masking** - Redact or transform sensitive data

### **Dynamic View Functions**

Unity Catalog provides functions to dynamically limit access:
* `current_user()` - Returns current user's email address
* `is_account_group_member()` - Check account-level group membership (recommended)
* `is_member()` - Check workspace-level group membership (legacy)

**Recommendation:** Don't grant users ability to read tables/views referenced in the dynamic view.

### **Dynamic View Examples**

**Example 1: Column-Level Permissions**```sql
-- Only auditors can see full email addresses
-- Others see domain only
CREATE VIEW sales_redacted AS
SELECT
  user_id,
  CASE
    WHEN is_account_group_member('auditors') THEN email
    ELSE 'REDACTED'
  END AS email,
  country,
  product,
  total
FROM sales_raw;
```**Example 2: Row-Level Permissions**```sql
-- Only managers can view transactions > $1,000,000
-- Other users see filtered results
CREATE VIEW sales_redacted AS
SELECT
  user_id,
  country,
  product,
  total
FROM sales_raw
WHERE
  CASE
    WHEN is_account_group_member('managers') THEN TRUE
    ELSE total <= 1000000
  END;
```**Example 3: Data Masking with Regular Expressions**```sql
-- Extract email domain for analysts
-- Show full email for auditors
CREATE VIEW sales_redacted AS
SELECT
  user_id,
  region,
  CASE
    WHEN is_account_group_member('auditors') THEN email
    ELSE regexp_extract(email, '^.*@(.*)$', 1)
  END AS email
FROM sales_raw;

### **Key Behaviors**

**Refresh as Owner:**
* When pipeline update refreshes materialized view or streaming table
* Row filter and column mask functions run with **pipeline owner's rights**
* Functions checking user context (`CURRENT_USER`, `IS_MEMBER`) use **pipeline owner's context**

**Query:**
* When querying materialized view or streaming table
* Functions checking user context use **invoker's user context**
* Enforces user-specific data security and access controls

**Full Refresh:**
* Materialized views over source tables with row filters/column masks always use **full refresh**
* Reprocesses all data with latest definitions
* Ensures security policies are evaluated with most up-to-date data

### **Managing in Pipelines**

* Row filters and column masks should be added, updated, or dropped through `CREATE OR REFRESH` statement
* See **[Lakeflow Spark Declarative Pipelines SQL language reference](https://docs.databricks.com/aws/en/ldp/developer/sql-ref/)** for syntax

### **USING COLUMNS Clause**

Use when masking function must reference:
* **Static parameters**
* **Other columns in the table**

**Syntax:**```sql

ALTER COLUMN <col_name> 
SET MASK <mask_func_name> 
USING COLUMNS (<column_list>);
```**How it works:**
* First parameter always maps to masked column
* Additional parameters supplied via `USING COLUMNS`
* Can use static values or column names from same table

## **Best Practices**

1. **Use is_account_group_member()** - For Unity Catalog (not is_member())
2. **Test thoroughly** - Verify filters/masks work as expected for all user groups
3. **Document functions** - Add comments explaining security logic
4. **Use mapping tables** - For complex access control requirements
5. **Consider performance** - Filters applied at scan time are efficient
6. **Drop before deleting** - Always DROP ROW FILTER/MASK before dropping function
7. **Use dynamic views** - For complex transformations or multi-table logic
8. **Limit direct table access** - Grant access to views, not underlying tables
9. **Combine approaches** - Use row filters + column masks + dynamic views together
10. **Monitor access** - Use audit logs to track who accesses sensitive data

## **Important Notes**

### **Execution Context**

* **Definer's rights** - All filters run with definer's rights
* **Exception** - Functions checking user context (`CURRENT_USER`, `IS_ACCOUNT_GROUP_MEMBER`) run as invoker

### **Performance Considerations**

* **Masks applied at scan time** - Before expressions, predicates, or ordering
* **Join behavior** - Joins use masked values for comparison
* **No negative impact** - Dynamic views with CASE statements have no performance penalty

### **Limitations**

* **One row filter per table** - Cannot apply multiple row filters
* **Must drop before delete** - Always drop filter/mask before dropping function
* **Type matching** - Mask function return type must match column type
* **Inaccessible state** - Table becomes inaccessible if function dropped before filter/mask

````sql
%python
from pyspark import pipelines as dp
from pyspark.sql.functions import *

# Step 1: Ingest raw data with Auto Loader
@dp.table(comment="Raw user data with PII")
def users_raw():
    return (
        spark.readStream
        .format("cloudFiles")
        .option("cloudFiles.format", "json")
        .option("cloudFiles.schemaLocation", "/checkpoint/users_raw")
        .load("/source/users/")
    )

# Step 2: Bronze layer - detect PII columns
@dp.table(
    comment="Bronze layer with PII detection",
    table_properties={"quality": "bronze"}
)
@dp.expect("valid_user_id", "user_id IS NOT NULL")
@dp.expect("valid_email_format", "email RLIKE '^[^@]+@[^@]+\\.[^@]+$'")
def users_bronze():
    return (
        spark.readStream.table("users_raw")
        .withColumn("ingestion_timestamp", current_timestamp())
        .withColumn("contains_pii", lit(True))
        .withColumn("pii_columns", array(lit("email"), lit("phone"), lit("ssn")))
    )

# Step 3: Silver layer - apply PII masking
@dp.table(
    comment="Silver layer with masked PII",
    table_properties={"quality": "silver"}
)
def users_silver():
    return (
        spark.readStream.table("users_bronze")
        .withColumn("email_masked", sha2(col("email"), 256))
        .withColumn("phone_masked", 
            concat(lit("***-***-"), substring(col("phone"), -4, 4)))
        .withColumn("ssn_masked", lit("***-**-****"))
        .select(
            "user_id",
            "username",
            "email_masked",
            "phone_masked",
            "ssn_masked",
            "registration_date",
            "ingestion_timestamp"
        )
    )

# Step 4: Gold layer - aggregated data without PII
@dp.table(
    comment="Gold layer - aggregated metrics",
    table_properties={"quality": "gold"}
)
def user_metrics_gold():
    return (
        spark.read.table("users_silver")
        .groupBy(date_trunc("day", col("registration_date")).alias("registration_day"))
        .agg(
            count("*").alias("user_count"),
            countDistinct("user_id").alias("unique_users")
        )
    )
````

````sql
-- Step 1: Raw data ingestion
CREATE OR REFRESH STREAMING TABLE users_raw
AS SELECT * FROM STREAM read_files('/source/users/', format => 'json');

-- Step 2: Bronze layer with PII detection
CREATE OR REFRESH STREAMING TABLE users_bronze (
  CONSTRAINT valid_user_id EXPECT (user_id IS NOT NULL),
  CONSTRAINT valid_email EXPECT (email RLIKE '^[^@]+@[^@]+\\.[^@]+$')
)
COMMENT "Bronze layer with PII detection"
TBLPROPERTIES ("quality" = "bronze")
AS SELECT 
  *,
  current_timestamp() AS ingestion_timestamp,
  true AS contains_pii,
  array('email', 'phone', 'ssn') AS pii_columns
FROM STREAM(users_raw);

-- Step 3: Create masking functions
CREATE FUNCTION mask_email(email STRING)
RETURN SHA2(email, 256);

CREATE FUNCTION mask_phone(phone STRING)
RETURN CONCAT('***-***-', RIGHT(phone, 4));

CREATE FUNCTION mask_ssn(ssn STRING)
RETURN '***-**-****';

-- Step 4: Silver layer with masked PII
CREATE OR REFRESH STREAMING TABLE users_silver (
  user_id INT,
  username STRING,
  email STRING MASK mask_email,
  phone STRING MASK mask_phone,
  ssn STRING MASK mask_ssn,
  registration_date STRING,
  ingestion_timestamp TIMESTAMP
)
COMMENT "Silver layer with masked PII"
TBLPROPERTIES ("quality" = "silver")
AS SELECT 
  user_id,
  username,
  email,
  phone,
  ssn,
  registration_date,
  ingestion_timestamp
FROM STREAM(users_bronze);

-- Step 5: Gold layer - aggregated without PII
CREATE OR REFRESH MATERIALIZED VIEW user_metrics_gold
COMMENT "Gold layer - aggregated metrics"
TBLPROPERTIES ("quality" = "gold")
AS SELECT
  DATE_TRUNC('day', registration_date) AS registration_day,
  COUNT(*) AS user_count,
  COUNT(DISTINCT user_id) AS unique_users
FROM users_silver
GROUP BY DATE_TRUNC('day', registration_date);
````

````sql
1. Rewrites all data files with deletion vectors
2. Creates new table version
3. Old files remain for time travel
4. Must run VACUUM to permanently delete old files

### **Complete Deletion Process**

**For Tables with Deletion Vectors:**```sql
-- Step 1: Delete records
DELETE FROM my_table WHERE user_id IN (SELECT user_id FROM gdpr_requests);

-- Step 2: Apply physical deletes
REORG TABLE my_table APPLY (PURGE);

-- Step 3: Wait for retention period (default 7 days)
-- This ensures time travel still works

-- Step 4: Permanently remove old files
VACUUM my_table RETAIN 168 HOURS;  -- 7 days
```**⚠️ Critical:** REORG creates new version. VACUUM retention period calculated from REORG completion time, not original DELETE time.

### **Streaming Tables and Materialized Views**

* **[Use streaming tables in Databricks SQL - Permanently delete records](https://docs.databricks.com/aws/en/ldp/dbsql/streaming/)** - Streaming table deletion

* **[Use materialized views in Databricks SQL - Permanently delete records](https://docs.databricks.com/aws/en/ldp/dbsql/materialized/)** - Materialized view deletion

**Process for Streaming Tables:**
1. **DELETE or UPDATE** records from streaming table
2. **REORG TABLE** with APPLY (PURGE)
3. **Wait** for retention period (default 7 days)
4. **REFRESH** streaming table
5. **Automatic VACUUM** runs within 24 hours

**Process for Materialized Views:**
1. **DELETE** data from source tables
2. **REORG TABLE** on materialized view with APPLY (PURGE)
3. **Wait** for retention period
4. **REFRESH** materialized view
5. **Automatic VACUUM** runs within 24 hours

## **GDPR/CCPA Compliance Implementation**

### **Blueprint for "Right to be Forgotten"**

**Architecture:**
1. **Control Table** - Track deletion requests
2. **Bronze Layer** - Delete PII first
3. **Silver Layer** - Propagate deletions
4. **Gold Layer** - Propagate deletions
5. **Upstream Sources** - Delete from Kafka, files, databases

### **Implementation Steps**

**Step 1: Create Control Table**```sql
-- Control table for deletion requests
CREATE TABLE gdpr_requests (
  user_id STRING,
  request_date TIMESTAMP,
  processed BOOLEAN DEFAULT FALSE,
  processed_date TIMESTAMP
);

-- Insert deletion request
INSERT INTO gdpr_requests (user_id, request_date)
VALUES ('user123', current_timestamp());
```**Step 2: Delete from Bronze Layer**```sql
-- Delete user data from bronze layer
DELETE FROM users_bronze 
WHERE user_id IN (
  SELECT user_id 
  FROM gdpr_requests 
  WHERE processed = FALSE
);

-- Delete user events from bronze layer
DELETE FROM events_bronze
WHERE user_id IN (
  SELECT user_id 
  FROM gdpr_requests 
  WHERE processed = FALSE
);
```**Step 3: Apply Physical Deletes (if using deletion vectors)**```sql
-- Purge soft-deletes
REORG TABLE users_bronze APPLY (PURGE);
REORG TABLE events_bronze APPLY (PURGE);
```**Step 4: Propagate to Silver Layer**```sql
-- Propagate deletions to silver layer
DELETE FROM users_silver 
WHERE user_id IN (SELECT user_id FROM gdpr_requests WHERE processed = FALSE);

DELETE FROM events_silver
WHERE user_id IN (SELECT user_id FROM gdpr_requests WHERE processed = FALSE);

-- Apply physical deletes
REORG TABLE users_silver APPLY (PURGE);
REORG TABLE events_silver APPLY (PURGE);
```**Step 5: Propagate to Gold Layer**```sql
-- For aggregated tables, may need to recalculate
-- Option 1: Delete and recalculate
DELETE FROM user_metrics_gold
WHERE user_id IN (SELECT user_id FROM gdpr_requests WHERE processed = FALSE);

-- Option 2: Use MERGE for upserts
MERGE INTO user_metrics_gold AS target
USING (
  SELECT user_id, COUNT(*) as event_count
  FROM events_silver
  WHERE user_id NOT IN (SELECT user_id FROM gdpr_requests)
  GROUP BY user_id
) AS source
ON target.user_id = source.user_id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```**Step 6: Mark as Processed**```sql
-- Update control table
UPDATE gdpr_requests
SET processed = TRUE,
    processed_date = current_timestamp()
WHERE processed = FALSE;
```**Step 7: Wait for Retention Period**```sql
-- Wait 7 days (or configured retention period)
-- Then run VACUUM to permanently delete
```**Step 8: Run VACUUM**```sql
-- Permanently remove deleted data
VACUUM users_bronze RETAIN 168 HOURS;
VACUUM events_bronze RETAIN 168 HOURS;
VACUUM users_silver RETAIN 168 HOURS;
VACUUM events_silver RETAIN 168 HOURS;
VACUUM user_metrics_gold RETAIN 168 HOURS;
```**Step 9: Delete from Upstream Sources**
* Delete from Kafka topics
* Delete from cloud storage
* Delete from external databases
* Delete from backup systems

## **Automated Data Purging Workflow**

### **Scheduled Job Implementation**

* **[Automating jobs with schedules and triggers](https://docs.databricks.com/aws/en/jobs/triggers/)** - Job automation

* **[Best practices for operational excellence - Automated workflows](https://docs.databricks.com/aws/en/lakehouse-architecture/operational-excellence/best-practices/)** - Workflow best practices

### **Create Scheduled Purging Job**

**Python Notebook:**```python
from pyspark.sql.functions import *
from datetime import datetime, timedelta

# Configuration
RETENTION_DAYS = 90  # Data older than 90 days will be purged
VACUUM_RETENTION_HOURS = 168  # 7 days

def purge_old_data():
    """Purge data older than retention period."""
    
    # Calculate cutoff date
    cutoff_date = (datetime.now() - timedelta(days=RETENTION_DAYS)).strftime('%Y-%m-%d')
    
    # Delete old records from each table
    tables_to_purge = [
        "catalog.schema.users",
        "catalog.schema.events",
        "catalog.schema.transactions"
    ]
    
    for table in tables_to_purge:
        print(f"Purging data from {table} older than {cutoff_date}")
        
        # Delete old records
        spark.sql(f"""
            DELETE FROM {table}
            WHERE date < '{cutoff_date}'
        """)
        
        # Apply physical deletes (if deletion vectors enabled)
        spark.sql(f"""
            REORG TABLE {table} 
            WHERE date < '{cutoff_date}'
            APPLY (PURGE)
        """)
        
        print(f"Completed purge for {table}")
    
    # Wait for retention period, then VACUUM
    # This should be a separate job scheduled 7+ days later
    
def vacuum_tables():
    """Run VACUUM to permanently delete old files."""
    
    tables_to_vacuum = [
        "catalog.schema.users",
        "catalog.schema.events",
        "catalog.schema.transactions"
    ]
    
    for table in tables_to_vacuum:
        print(f"Running VACUUM on {table}")
        
        spark.sql(f"""
            VACUUM {table} RETAIN {VACUUM_RETENTION_HOURS} HOURS
        """)
        
        print(f"Completed VACUUM for {table}")

# Run purge
purge_old_data()
```**SQL Notebook:**```sql
-- Configuration
DECLARE cutoff_date = current_date() - INTERVAL '90' DAY;

-- Delete old records from users table
DELETE FROM catalog.schema.users
WHERE registration_date < cutoff_date;

-- Apply physical deletes
REORG TABLE catalog.schema.users 
WHERE registration_date < cutoff_date
APPLY (PURGE);

-- Delete old records from events table
DELETE FROM catalog.schema.events
WHERE event_date < cutoff_date;

-- Apply physical deletes
REORG TABLE catalog.schema.events 
WHERE event_date < cutoff_date
APPLY (PURGE);

-- Note: Schedule separate job 7+ days later to run VACUUM
```### **Create Scheduled Job**

**Using Databricks UI:**
1. Create notebook with purging logic
2. Click **Schedule** button at top of editor
3. Choose **Add schedule**
4. Configure schedule:
   * **Daily** - Run every day at specific time
   * **Weekly** - Run once per week
   * **Monthly** - Run once per month
   * **Custom** - Use Quartz cron expression
5. Set notifications for job completion/failure
6. Click **Create** to save job

**Using Databricks CLI:**```bash
# Create job configuration file
cat > purge_job.json << EOF
{
  "name": "Data Purging Job",
  "tasks": [
    {
      "task_key": "purge_old_data",
      "notebook_task": {
        "notebook_path": "/Users/user@databricks.com/purge_data",
        "source": "WORKSPACE"
      },
      "job_cluster_key": "purge_cluster"
    },
    {
      "task_key": "vacuum_tables",
      "depends_on": [{"task_key": "purge_old_data"}],
      "notebook_task": {
        "notebook_path": "/Users/user@databricks.com/vacuum_tables",
        "source": "WORKSPACE"
      },
      "job_cluster_key": "purge_cluster"
    }
  ],
  "job_clusters": [
    {
      "job_cluster_key": "purge_cluster",
      "new_cluster": {
        "spark_version": "14.3.x-scala2.12",
        "node_type_id": "i3.xlarge",
        "num_workers": 2
      }
    }
  ],
  "schedule": {
    "quartz_cron_expression": "0 0 2 * * ?",
    "timezone_id": "America/Los_Angeles",
    "pause_status": "UNPAUSED"
  }
}
EOF

# Create job
databricks jobs create --json-file purge_job.json
```## **Predictive Optimization for Automatic Maintenance**

### **Primary Documentation**

* **[Predictive optimization for Unity Catalog managed tables](https://docs.databricks.com/aws/en/optimizations/predictive-optimization/)** - Automatic maintenance

### **What Predictive Optimization Does**

**Automatically:**
* Identifies tables needing maintenance
* Queues ANALYZE, OPTIMIZE, and VACUUM operations
* Collects statistics when data is written
* Runs on serverless compute
* Eliminates manual maintenance burden

**For Lakeflow Pipelines:**
* Intelligently maintains streaming tables and materialized views
* Based on usage patterns
* Automatic maintenance within 24 hours of updates

### **Enable Predictive Optimization**

**Account Level:**```sql
-- Enable for entire account (requires account admin)
ALTER ACCOUNT SET PREDICTIVE_OPTIMIZATION = ON;
```**Catalog Level:**```sql
-- Enable for specific catalog (requires catalog owner)
ALTER CATALOG my_catalog 
SET TBLPROPERTIES ('delta.enablePredictiveOptimization' = 'true');
```**Schema Level:**```sql
-- Enable for specific schema (requires schema owner)
ALTER SCHEMA my_catalog.my_schema
SET TBLPROPERTIES ('delta.enablePredictiveOptimization' = 'true');
```**Table Level:**```sql
-- Enable for specific table
ALTER TABLE my_catalog.my_schema.my_table
SET TBLPROPERTIES ('delta.enablePredictiveOptimization' = 'true');
```### **Monitor Predictive Optimization**

* **[Predictive optimization system table reference](https://docs.databricks.com/aws/en/admin/system-tables/predictive-optimization/)** - Tracking operations

**Query Operations:**```sql
-- View predictive optimization operations
SELECT *
FROM system.storage.predictive_optimization_operations_history
WHERE table_name = 'my_catalog.my_schema.my_table'
ORDER BY start_time DESC
LIMIT 100;

-- View costs
SELECT 
  table_name,
  operation_type,
  SUM(usage_quantity) as total_usage,
  SUM(usage_quantity * list_price) as estimated_cost
FROM system.storage.predictive_optimization_operations_history
WHERE start_time >= current_date() - INTERVAL '30' DAY
GROUP BY table_name, operation_type
ORDER BY estimated_cost DESC;
```### **Limitations**

* Not available in all regions
* Only Unity Catalog managed tables
* External tables not supported
* Delta Sharing recipient tables not supported
* For tables with retention < 7 days, VACUUM runs with 7-day retention

## **DELETE Operations**

### **Primary Documentation**

* **[DELETE FROM](https://docs.databricks.com/aws/en/sql/language-manual/delta-delete-from/)** - DELETE syntax

* **[Tutorial: Create and manage Delta Lake tables - Delete from a table](https://docs.databricks.com/aws/en/delta/tutorial/)** - Examples

### **DELETE Syntax**```sql
-- Basic DELETE
DELETE FROM table_name WHERE predicate;

-- Delete all rows
DELETE FROM table_name;

-- Delete with subquery
DELETE FROM users 
WHERE user_id IN (SELECT user_id FROM gdpr_requests);

-- Delete with date filter
DELETE FROM events
WHERE event_date < current_date() - INTERVAL '90' DAY;
```**Python API:**```python
from delta.tables import *
from pyspark.sql.functions import *

# Load Delta table
deltaTable = DeltaTable.forName(spark, "catalog.schema.my_table")

# Delete using SQL-formatted string
deltaTable.delete("age < '18'")

# Delete using Spark SQL functions
deltaTable.delete(col('age') < '21')

# Delete with complex predicate
deltaTable.delete(
    (col('event_date') < current_date() - expr("INTERVAL 90 DAYS"))
)
```**⚠️ Important:** Deletion removes data from latest version but doesn't remove from physical storage until VACUUM runs.

## **MERGE Operations for Data Retention**

### **Primary Documentation**

* **[MERGE INTO](https://docs.databricks.com/aws/en/sql/language-manual/delta-merge-into/)** - MERGE syntax and examples

### **MERGE for Conditional Deletion**```sql
-- Delete unmatched rows (data not in source)
MERGE INTO target USING source
ON target.key = source.key
WHEN NOT MATCHED BY SOURCE THEN DELETE;

-- Conditional deletion based on flag
MERGE INTO target USING source
ON target.key = source.key
WHEN NOT MATCHED BY SOURCE AND target.marked_for_deletion THEN DELETE
WHEN NOT MATCHED BY SOURCE THEN UPDATE SET target.value = DEFAULT;

-- Delete matched rows conditionally
MERGE INTO target USING source
ON target.key = source.key
WHEN MATCHED AND target.marked_for_deletion THEN DELETE
WHEN MATCHED THEN UPDATE SET *;
```### **MERGE for SCD Type 2 (Slowly Changing Dimensions)**

**Track Historical Changes:**```sql
-- SCD Type 2 implementation
MERGE INTO customer_history AS target
USING customer_updates AS source
ON target.customer_id = source.customer_id 
   AND target.is_current = TRUE
WHEN MATCHED AND target.hash != source.hash THEN
  UPDATE SET 
    is_current = FALSE,
    end_date = current_date()
WHEN NOT MATCHED THEN
  INSERT (customer_id, name, email, start_date, end_date, is_current, hash)
  VALUES (source.customer_id, source.name, source.email, 
          current_date(), NULL, TRUE, source.hash);

-- Insert new version of changed records
INSERT INTO customer_history
SELECT 
  customer_id, name, email,
  current_date() as start_date,
  NULL as end_date,
  TRUE as is_current,
  hash
FROM customer_updates
WHERE customer_id IN (
  SELECT customer_id FROM customer_history 
  WHERE is_current = FALSE 
  AND end_date = current_date()
);
```## **Change Data Feed for Compliance Tracking**

### **Primary Documentation**

* **[Use Delta Lake change data feed on Databricks](https://docs.databricks.com/aws/en/delta/delta-change-data-feed/)** - Complete CDF guide

### **What is Change Data Feed?**

**Tracks:**
* Row-level changes between table versions
* Insert, delete, and update operations
* Row data + metadata about change type

**Use Cases:**
* **ETL pipelines** - Incrementally process only changed rows
* **Audit trails** - Track data modifications for compliance
* **Data replication** - Synchronize changes to downstream systems

### **Enable Change Data Feed**```sql
-- Enable on new table
CREATE TABLE my_table (columns...)
TBLPROPERTIES (delta.enableChangeDataFeed = true);

-- Enable on existing table
ALTER TABLE my_table 
SET TBLPROPERTIES (delta.enableChangeDataFeed = true);
```### **Query Change Data Feed**

**SQL:**```sql
-- Query changes between versions
SELECT * FROM table_changes('my_table', 2, 5);

-- Query changes since timestamp
SELECT * FROM table_changes('my_table', '2024-01-01');

-- Query changes between timestamps
SELECT * FROM table_changes('my_table', '2024-01-01', '2024-01-31');
```**Streaming:**```python
# Read change data feed as stream
(spark.readStream
    .option("readChangeFeed", "true")
    .table("my_table")
    .writeStream
    .format("delta")
    .outputMode("append")
    .option("checkpointLocation", "/checkpoint/path")
    .table("my_table_changes")
)
```**Change Data Columns:**
* `_change_type` - insert, update_preimage, update_postimage, delete
* `_commit_version` - Table version of change
* `_commit_timestamp` - Timestamp of change

### **Retention Considerations**

* Change data feed records are **transient**
* Only accessible within retention window
* Removed when table versions are removed
* Follow table's retention policy
* VACUUM deletes change data files

## **Audit and Monitoring**

### **DESCRIBE HISTORY**

* **[DESCRIBE HISTORY](https://docs.databricks.com/aws/en/sql/language-manual/delta-describe-history/)** - Table history tracking

**Syntax:**```sql
-- Get full history
DESCRIBE HISTORY table_name;

-- Get last operation only
DESCRIBE HISTORY table_name LIMIT 1;

-- Get history for specific time range
DESCRIBE HISTORY table_name 
WHERE timestamp >= '2024-01-01';
```**History Schema:**

| Column | Type | Description |
|--------|------|-------------|
| version | long | Table version |
| timestamp | timestamp | When version was committed |
| userId | string | ID of user who ran operation |
| userName | string | Name of user |
| operation | string | Operation name (DELETE, UPDATE, MERGE, etc.) |
| operationParameters | map | Operation parameters (predicates, etc.) |
| operationMetrics | map | Metrics (rows/files modified) |
| job | struct | Job details |
| notebook | struct | Notebook details |
| clusterId | string | Cluster ID |

**Example Output:**```
+-------+-------------------+------+--------+---------+--------------------+
|version|     timestamp     |userId|userName|operation| operationParameters|
+-------+-------------------+------+--------+---------+--------------------+
|     5|2019-07-29 14:07:47|   ###|     ###|   DELETE|[predicate -> [(...|
|     4|2019-07-29 14:07:41|   ###|     ###|   UPDATE|[predicate -> (id...|
|     3|2019-07-29 14:07:29|   ###|     ###|   DELETE|[predicate -> [(...|
+-------+-------------------+------+--------+---------+--------------------+
```### **System Tables for Monitoring**

* **[Monitor account activity with system tables](https://docs.databricks.com/aws/en/admin/system-tables/index/)** - System tables overview

* **[Audit log system table reference](https://docs.databricks.com/aws/en/admin/system-tables/audit-logs/)** - Audit logs

**Query Audit Logs:**
````
