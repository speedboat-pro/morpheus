Here's the generally accepted guidance from Databricks architecture best practices:

- Silver Layer — enterprise-conforming, cleansed, joined data:
    - Inmon (3NF): Fits silver when the goal is a normalized, enterprise-wide "single source of truth" — cleansed and conformed but not aggregated for a specific business use case.
    - Linstedt (Data Vault): Hubs, Links, and Satellites naturally belong in silver — they prioritize auditability, historization, and flexibility without business-specific transformation.
- Gold Layer — business-level, consumption-ready data:
    - Kimball (Dimensional/Star Schema): Fact and dimension tables optimized for specific analytic consumption (dashboards, reports, ML features) belong in gold.
    - Inmon or Data Vault structures also land in gold if they've been further aggregated or shaped for a specific business domain or use case.

| Criteria | Silver | Gold |
| --- | --- | --- |
| Purpose | Conform & cleanse | Business consumption |
| Audience | Multiple downstream teams | Specific business unit/use case |
| Granularity | Atomic/transactional | Often aggregated or denormalized |
| Modeling style | 3NF, Data Vault | Star schema, wide/flat tables, aggregates |
| Reusability | High (many gold tables derive from it) | Lower (purpose-built) |
