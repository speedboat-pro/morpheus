```python
import pandas as pd
from sklearn.datasets import load_iris

def sample_iris_data():
  iris = load_iris()
  iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
  iris_df['quality'] = (iris.target == 2).astype(int)
  return iris_df
```

```python
import mlflow

# REQUIRED: add evaluation dataset and target here
data = sample_iris_data()
target = "quality"
# REQUIRED: add model type here (e.g. "regressor", "databricks-agent", etc.)
model_type = "regressor"

model_name = dbutils.widgets.get("model_name")
model_version = dbutils.widgets.get("model_version")
model_uri = "models:/" + model_name + "/" + model_version 
# can also fetch model ID and use that for URI instead as described below

with mlflow.start_run(run_name="evaluation") as run:
  eval_result = mlflow.models.evaluate(
    model=model_uri,
    data=data,
    targets=target,
    model_type=model_type
  )

  # mae = eval_result.metrics.get("mean_absolute_error")
  # print(f"MAE: {mae}")

  # if mae is not None and mae > 500:
  #   mlflow.set_tag("OK", "approved")
  #   print("Tag 'OK' set to 'approved'")
  # else:
  #   print(f"MAE ({mae}) is not above 500 — tag not set")
```
