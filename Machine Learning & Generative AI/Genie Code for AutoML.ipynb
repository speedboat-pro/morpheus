##Classification Prompt

[1.1a Demo]($../../machine-learning-model-development-en_us-2.1.6/M01 - Model Development Workflow/1.1a Demo - Supervised Learning)

```



You are a data science agent working in a Databricks notebook.  
We have a Spark DataFrame called `train_df_cls` and `test_df_cls`.  
Target column: `is_high_quality` (binary classification).  
Exclude identifier column: `wine_id`.  

Build a baseline classification workflow with these requirements:
- Use Spark ML where possible (LogisticRegression and RandomForestClassifier are fine)
- Handle categorical features appropriately (StringIndexer + OneHotEncoder)
- Assemble features using VectorAssembler
- Evaluate on the test set using: Accuracy, Precision, Recall, F1-score, ROC-AUC
- Log each model as an MLflow run in the current experiment:
  - Log parameters and metrics
  - Log the model artifact
  - IMPORTANT: When logging the model, include an MLflow model signature (required for Unity Catalog)
    - Use `mlflow.models.signature.infer_signature()`
    - Signature inputs should be a pandas sample of the original feature columns
      (exclude `is_high_quality` and `wine_id`)
    - Signature outputs should include at least the `prediction` column

- Pick the best model based on F1-score (higher is better) and return:
  - the best run_id
  - the best metrics
  - code to load the best model for inference

Keep the code simple and readable for learners.
```
